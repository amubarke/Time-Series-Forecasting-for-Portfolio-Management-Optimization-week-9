```python
import pandas as pd

# Example: filter last year for backtesting
backtest_start = "2025-01-01"
backtest_end   = "2026-01-01"

tsla_bt = tsla_prices[backtest_start:backtest_end]
spy_bt  = spy_prices[backtest_start:backtest_end]
bnd_bt  = bnd_prices[backtest_start:backtest_end]

# Combine into a DataFrame
prices_bt = pd.concat([tsla_bt, spy_bt, bnd_bt], axis=1)
prices_bt.columns = ['TSLA', 'SPY', 'BND']
```

```python
benchmark_weights = pd.Series({'TSLA': 0, 'SPY': 0.6, 'BND': 0.4})
```

```python
strategy_weights = pd.Series({
    'TSLA': 0.5,
    'SPY': 0.3,
    'BND': 0.2
})
```

```python
# Daily returns
returns_bt = prices_bt.pct_change().dropna()

# Strategy cumulative returns
strategy_daily_returns = (returns_bt * strategy_weights).sum(axis=1)
strategy_cum_returns = (1 + strategy_daily_returns).cumprod()

# Benchmark cumulative returns
benchmark_daily_returns = (returns_bt * benchmark_weights).sum(axis=1)
benchmark_cum_returns = (1 + benchmark_daily_returns).cumprod()
```

```python
import numpy as np

def compute_metrics(cum_returns, risk_free_rate=0.04):
    daily_returns = cum_returns.pct_change().dropna()
    total_return = cum_returns[-1] / cum_returns[0] - 1
    annualized_return = (1 + total_return) ** (252/len(daily_returns)) - 1
    annualized_vol = daily_returns.std() * np.sqrt(252)
    sharpe_ratio = (annualized_return - risk_free_rate) / annualized_vol
    drawdown = (cum_returns / cum_returns.cummax() - 1).min()
    
    return {
        "Total Return": total_return,
        "Annualized Return": annualized_return,
        "Annualized Volatility": annualized_vol,
        "Sharpe Ratio": sharpe_ratio,
        "Max Drawdown": drawdown
    }

strategy_metrics = compute_metrics(strategy_cum_returns)
benchmark_metrics = compute_metrics(benchmark_cum_returns)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(strategy_cum_returns, label="Strategy Portfolio")
plt.plot(benchmark_cum_returns, label="Benchmark (60% SPY / 40% BND)")
plt.title("Backtesting: Strategy vs Benchmark")
plt.xlabel("Date")
plt.ylabel("Cumulative Returns")
plt.legend()
plt.show()
```
